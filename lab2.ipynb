function:  $$sin(x^2)+cos(x) $$ \\
range of acceptable values: $$[-5; 5]$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
from torch.autograd import Variable
```

```python
x = np.linspace(-5, 5, 200)
y = np.sin(x ** 2) + np.cos(x)
```

```python
plt.plot(x, y)
plt.title('function: sin(x^2) + cos(x)')
plt.show()
```
